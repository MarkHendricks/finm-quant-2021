```python
import pandas as pd
import numpy as np
import statsmodels.api as sm

path_to_data_file = 'multi_asset_returns.xlsx'
```

# Load the data from HW #1

```python
retsx = pd.read_excel(path_to_data_file)
retsx.set_index('Date',inplace=True)
retsx
```

# Try a univariate regression, (just one X variable)

```python
X = sm.add_constant(retsx['SPY'])
y = retsx['DBC']
model_uni = sm.OLS(y, X).fit()
model_uni.summary()
```

# Try a multivariate regression

Does the beta on SPY change when we add a second regressor, compared to the single variable SPY?

```python
X = sm.add_constant(retsx[['SPY','EFA']])
y = retsx['DBC']
model_multi = sm.OLS(y, X).fit()
model_multi.summary()
```
